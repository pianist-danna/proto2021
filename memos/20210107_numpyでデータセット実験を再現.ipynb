```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def vis(dataset):
    x = []
    for item in dataset:
        x.append(item)
    plt.plot(x)
    plt.legend("data sequence")
    plt.title("Dataset Output")
    plt.show()

#Generatorオブジェクトへの変換関数
def yield_dataset(data):
    for i in range(len(data)):
        yield data[i]

x = np.arange(100)
dataset = yield_dataset(x)
vis(dataset)
```

### リピートする
`np.tile`を使うとリピートできる

```python
x = np.arange(100)
x = np.tile(x,3)
dataset = yield_dataset(x)
vis(dataset)
```

### シャッフルする
numpyでは`np.random.default_rng`をインスタンス化し、そこに配列を通すことでランダマイズされる  
ランダマイズの幅は指定できない(する必要性も薄いが…)

`np.random.shuffle()`に通してもランダマイズできるが、こちらのほうが高速

```python
x = np.arange(100)
x = np.tile(x,3)
rng = np.random.default_rng()
rng.shuffle(x)
dataset = yield_dataset(x)
vis(dataset)
```
